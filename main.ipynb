```python
import pandas as pd
```

```python
manga_raw = pd.read_csv('./data/manga_raw.csv', dtype={'manga_favorites': 'object'})

df = manga_raw.copy()
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['manga_favorites'] = df['manga_favorites'].str.replace(r'[,]', '', regex=True).astype('int')
```

```python
df['manga_members'] = df['manga_members'].str.replace(r'[,]', '', regex=True).astype('int')
```

```python
df['manga_popularity'] = df['manga_popularity'].str.replace(r'[#]', '', regex=True).astype('int')
```

```python
df['manga_ranked'] = df['manga_ranked'].str.extract(r'(\d+)22').astype('float')
```

```python
df['manga_rating'] = df['manga_score'].str.extract(r'(\d+.\d+)').astype('float')
```

```python
df['manga_rating_by_user'] = df['manga_score'].str.extract(r'(\d+,\d+)').replace(r'[,]', '', regex=True)

df['manga_rating_by_user_div'] = 10 ** (df['manga_rating_by_user'].str.len() // 2)

df['manga_rating_by_user'] = df['manga_rating_by_user'].astype('float')

df['manga_rating_by_user'] = df['manga_rating_by_user'] // df['manga_rating_by_user_div']

df.drop(['manga_score', 'manga_rating_by_user_div'], axis=1, inplace=True)
```

```python
df['manga_volume'] = pd.to_numeric(df['manga_volume'], errors='coerce')
```

```python
df['manga_chapters'] = pd.to_numeric(df['manga_chapters'], errors='coerce')
```

```python
published_from_to = df['manga_published'].str.split(r'\s+to\s+', n=1, expand=True)

df['manga_published_from'] = pd.to_datetime(published_from_to[0], format='%b %-d, %Y', errors='coerce')

df['manga_published_from'] = df['manga_published_from'].fillna(
    pd.to_datetime(published_from_to[0], format='%b %Y', errors='coerce')
)

df['manga_published_from'] = df['manga_published_from'].fillna(
    pd.to_datetime(published_from_to[0], format='%Y', errors='coerce')
)

df['manga_published_to'] = pd.to_datetime(published_from_to[1], format='%b %-d, %Y', errors='coerce')

df['manga_published_to'] = df['manga_published_to'].fillna(
    pd.to_datetime(published_from_to[1], format='%b %Y', errors='coerce')
)

df['manga_published_to'] = df['manga_published_to'].fillna(
    pd.to_datetime(published_from_to[1], format='%Y', errors='coerce')
)

df.drop('manga_published', axis=1, inplace=True)
```

```python
genre_df = df['manga_genre'].str.split(',', expand=True)

col = []

for idx in range(1, genre_df.shape[1] + 1):
    col.append(f'manga_genre_{idx}')

genre_df.columns = col

genre_df = genre_df.replace(r'^(.+)\1$', r'\1', regex=True)

df = df.join(genre_df, how='left')

df.drop('manga_genre', axis=1, inplace=True)
```

```python
theme_df = df['manga_theme'].str.split(',', expand=True)

col = []

for idx in range(1, theme_df.shape[1] + 1):
    col.append(f'manga_theme_{idx}')

theme_df.columns = col

theme_df = theme_df.replace(r'^(.+)\1$', r'\1', regex=True)

df = df.join(theme_df, how='left')

df.drop('manga_theme', axis=1, inplace=True)
```

```python
demographic_df = df['manga_demographic'].str.split(',', expand=True)

col = []

for idx in range(1, demographic_df.shape[1] + 1):
    col.append(f'manga_demographic_{idx}')

demographic_df.columns = col

demographic_df = demographic_df.replace(r'^(.+)\1$', r'\1', regex=True)

df = df.join(demographic_df, how='left')

df.drop('manga_demographic', axis=1, inplace=True)
```

```python
author_df = df[['manga_author']].copy()

story_art = author_df['manga_author'].str.extract(r'([^()]+)\(Story\s*&\s*Art\)')
story_only = author_df['manga_author'].str.extract(r'([^()]+)\(Story\)')
art_only = author_df['manga_author'].str.extract(r'([^()]+)\(Art\)')

author_df['manga_author_story'] = story_art.fillna(story_only)
author_df['manga_author_art'] = story_art.fillna(art_only)


def fix_name(s: str):
    if s is None or pd.isna(s):
        return pd.NA
    s = str(s).strip(',')
    parts = [p.strip() for p in s.split(',')]
    if len(parts) >= 2:
        return ' '.join(parts[1:] + [parts[0]])
    return parts[0] if parts else pd.NA


author_df['manga_author_story'] = author_df['manga_author_story'].apply(fix_name)
author_df['manga_author_art'] = author_df['manga_author_art'].apply(fix_name)

df = df.join(author_df.drop('manga_author', axis=1), how='left').drop('manga_author', axis=1)
```

```python
df.drop_duplicates(ignore_index=True, inplace=True)

df.info()
```
